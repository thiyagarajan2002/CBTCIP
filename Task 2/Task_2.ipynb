```python
import pandas as pd
import numpy as np

df1 = pd.read_csv('/content/Unemployment in India.csv')
df2 = pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')
```

```python
frames=[df1,df2]
df=pd.concat(frames)
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Region']= df['Region'].fillna(df['Region'].mode()[0])
df['Date']= df['Date'].fillna(df['Date'].mode()[0])
df['Frequency']= df['Frequency'].fillna(df['Frequency'].mode()[0])
df['Estimated Unemployment Rate (%)'] = df['Estimated Unemployment Rate (%)'].fillna(df['Estimated Unemployment Rate (%)'].mean())
df['Estimated Employed'] = df['Estimated Employed'].fillna(df['Estimated Employed'].mean())
df['Estimated Labour Participation Rate (%)']= df['Estimated Labour Participation Rate (%)'].fillna(df['Estimated Labour Participation Rate (%)'].mean())
df['Area']= df['Area'].fillna(df['Area'].mode()[0])
df['Region.1']= df['Region.1'].fillna(df['Region.1'].mode()[0])
df['Latitude']= df['Latitude'].fillna(df['Latitude'].mean())
df['Longitude']= df['Longitude'].fillna(df['Longitude'].mean())
df.isnull().sum()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
plt.figure(figsize=(500, 5.2))
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['Date']
  ys = series['Estimated Unemployment Rate (%)']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = df.sort_values('Date', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('Date')
_ = plt.ylabel('Estimated Unemployment Rate (%)')
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Region']=le.fit_transform(df['Region'])
df['Region.1']=le.fit_transform(df['Region.1'])
df['Date']=le.fit_transform(df['Date'])
df['Frequency']=le.fit_transform(df['Frequency'])
df['Area']=le.fit_transform(df['Area'])
df.head()
```

```python
x=df.drop('Estimated Unemployment Rate (%)',axis=1)
y=df['Estimated Unemployment Rate (%)']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
r2_score_value=r2_score(y_test,y_pred)
print(r2_score_value)
```
