```python
import pandas as pd
import numpy as np
```

```python
data_frame=pd.read_csv('/content/Iris Flower - Iris.csv')
```

```python
data_frame.head()
```

```python
data_frame.isnull().sum()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

data_frame['Species']= le.fit_transform(data_frame['Species'])
```

```python
data_frame.head()
```

```python
x=data_frame.drop(['Species','Id'],axis=1)
y=data_frame['Species']
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y , random_state=104, test_size=0.25)
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
#K Neighbors Classifier
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_pred=knn.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Mean absolute error:",mean_absolute_error(y_test, y_pred))
print("Mean squared error:",mean_squared_error(y_test, y_pred))
```

```python
import pickle
with open('model.pkl', 'wb') as file:
    pickle.dump(knn, file)
```

```python
with open('model.pkl', 'rb') as file:
    loaded_model = pickle.load(file)

new_data=pd.DataFrame([[4.6,3.1,1.5,0.2]],columns=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'])
predictions = loaded_model.predict(new_data)
print(predictions)
```
